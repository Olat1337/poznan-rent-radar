# 🏡 Poznań Rent Predictor: Project Overview
**Author:** Kyrylo Kudrevych

Welcome to the Poznań Rent Predictor project! 👋

### 🎯 The Problem
Finding an apartment can be stressful, and knowing if a listing is fairly priced is even harder. The goal of this project is to use real, scraped data from the Poznań real estate market to build a Machine Learning AI that can predict the fair monthly cost of any apartment based on its features (like area, floor, and location).

### 🗂️ Project Structure
To keep the code clean and easy to follow, I have split the entire Data Science workflow across four specific notebooks:

* **`1_EDA_and_Cleaning.ipynb`:** 🧹
  * *What it does:* Takes messy, raw text data scraped from the internet, cleans out extreme outliers (like typos or fake listings), and converts text categories into numbers.
* **`2_Finding_best_hyperparams.ipynb`:** 🎛️
  * *What it does:* Uses an optimization robot (`Optuna`) to test hundreds of settings for our algorithms to ensure they are tuned perfectly for our specific data.
* **`3_ML_training.ipynb`:** 🧠
  * *What it does:* Safely splits the data, trains 4 different tree-based models (Random Forest, LightGBM, XGBoost, CatBoost), and tests a combined "Ensemble" approach.
* **`4_Conclusion_Evaluation_and_Feature_Importance.ipynb`:** 🔎
  * *What it does:* Opens up our winning model to see *why* it makes its decisions and adds a safety feature to protect our final User Interface (UI).

### 🏆 Final Results
* **The Winner:** **CatBoost** 🥇
* **Accuracy:** We achieved a final Mean Absolute Error (MAE) of **~307 PLN**. This means that, on average, our model can predict the true total monthly cost of a Poznań apartment within about 300 PLN of reality!
* **The Solution:** We successfully saved the trained model (`catboost_housing_model.pkl`).
